# Softmax

cf. [Quiz: Softmax](https://classroom.udacity.com/courses/ud730/lessons/6370362152/concepts/63815621490923)

```python
"""Softmax"""
```

```python
scores = [3.0, 1.0, 0.2]
```

```python
import numpy as np
```

```python
def softmax(x):
    """Compute softmax values for x."""
    y = np.exp(x)/np.sum(np.exp(x), axis=0)
    return y
```

```python
print(softmax(scores))
```

```python
# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0,0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
```

```python
plt.plot(x, softmax(scores).T, linewidth=2)
```

```python
plt.show()
```

# Numerical Stability
EY : 20170126 Numerical Stability!!!  
cf. [Quiz: Numerical Stability](https://classroom.udacity.com/courses/ud730/lessons/6370362152/concepts/71235296110923#)

```python
N = 1000000000
a = 0.000001
ITER_MAX = 1000000
```

```python
for iter in range(ITER_MAX):
    N += a
```

```python
print( N - 1000000000 )
```

```python
from decimal import Decimal
```

```python
N_Decimal = Decimal(N)
a_Decimal = Decimal(a)
for iter in range(ITER_MAX):
    N_Decimal += a_Decimal
```

```python
print( N_Decimal - Decimal( N))
```
